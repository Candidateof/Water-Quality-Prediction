```python
pollutants = ['O2', 'NO3', 'NO2', 'SO4', 'PO4']
```

### WEEK 2 - 23 JUNE 2025

```python
# Drop the missing values
df = df.dropna(subset=pollutants)
df.head()
```

```python
df.isnull().sum()
```

```python
# feature and target selection - Feature - independent variable and target dependent variable 
# When we are using a column name (feature), then double square brackets are used [[]], for rest []
X = df[['id', 'year']]
y = df[pollutants]
# If pollutants were a actual column name (feature), only then the double square brackets are applied.
```

```python
# Encoding - onehotencoder - 22 stations - 1 - 1
# convert categorical data into numerical format
X_encoded = pd.get_dummies(X, columns=['id'], drop_first=True)
```

```python
# Train, Test and Split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
    # Feature, Target, Test, Random state
    # To fixate splitting, Random state is used
)
```

```python
# Train the model 
# Without MultiOutputRegressor, RandomForestRegressor handle one column as output
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
model.fit(X_train, y_train)
```

```python
# Evaluate model 
y_pred = model.predict(X_test)
```

```python
print("Model Performance on the Test Data:")
# Evaluate model's performance for each pollutant (target variable) using MSE and R2 score
for i, pollutant in enumerate(pollutants):
    print(f'{pollutant}:')
    print('    MSE:', mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
    print('    R2:', r2_score(y_test.iloc[:, i], y_pred[:, i]))
    print()
```

```python
station_id = '22'
year_input = 2024

input_data = pd.DataFrame({'year': [year_input], 'id': [station_id]})
input_encoded = pd.get_dummies(input_data, columns=['id'])

# Align with training feature columns
missing_cols = set(X_encoded.columns) - set(input_encoded.columns)
for col in missing_cols:
    input_encoded[col] = 0
input_encoded = input_encoded[X_encoded.columns]  # reorder columns

# Predict pollutants
predicted_pollutants = model.predict(input_encoded)[0]

print(f"\nPredicted pollutant levels for station '{station_id}' in {year_input}:")
for p, val in zip(pollutants, predicted_pollutants):
    print(f"  {p}: {val:.2f}")
```
